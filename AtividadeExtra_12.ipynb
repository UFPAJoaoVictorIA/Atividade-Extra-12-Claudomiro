### 📑 Descrição do Exercício
Este notebook implementa a simulação de uma memória RAM baseada nos conceitos apresentados nos slides: **Simulação de Gravação e Leitura de Dados**.

### 🚀 Passo a Passo do Funcionamento

1.  **Inicialização:** A memória é criada com **16 endereços** de 4 bits (de `0000` a `1111`). Inicialmente, todos os endereços estão "limpos" (contendo `00000000`).
2.  **Gravação (Write):** Permite inserir uma cadeia de bits (8 bits) em um endereço específico fornecido em formato binário.
3.  **Leitura (Read):** Recupera o dado armazenado em um endereço específico. Se o endereço não foi gravado, ele retornará o valor padrão inicial.
4.  **Listagem (List All):** Exibe o estado atual de toda a grade de memória, permitindo auditar o que foi gravado.

---

```python
class RAM:
    def __init__(self, size=16):
        # Inicializa a memória com zeros (limpa), conforme o slide 5
        self.size = size
        self.memory = {format(i, '04b'): "00000000" for i in range(size)}

    def write(self, address, data):
        """Grava dados em um endereço binário."""
        if address in self.memory:
            self.memory[address] = data
            print(f"Gravado: {data} no endereço {address}")
        else:
            print("Erro: Endereço inválido.")

    def read(self, address):
        """Lê um único endereço de memória (Slide 11/12)."""
        data = self.memory.get(address, "Endereço inexistente")
        print(f"Leitura do endereço {address}: {data}")
        return data

    def list_all(self):
        """Lista todos os endereços preenchidos (Slide 13/16)."""
        print("\n--- LISTA DA MEMÓRIA ---")
        for addr, value in self.memory.items():
            print(f"Endereço: {addr} | Valor: {value}")
        print("------------------------\n")

# --- Demonstração baseada nos slides ---

# 1. Instanciar o simulador
simulador = RAM()

# 2. Mostrar memória limpa (Slide 4/5)
simulador.list_all()

# 3. Realizar algumas gravações (Slides 6-10)
simulador.write("0000", "10101010")
simulador.write("0001", "11110000")
simulador.write("1111", "00001111")

# 4. Leitura individual (Slide 12)
simulador.read("0000")
simulador.read("0100") # Endereço zerado

# 5. Listar memória total gravada (Slide 16)
simulador.list_all()
```
